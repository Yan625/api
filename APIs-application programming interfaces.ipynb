# API

1.Why use API: cause data is dynamic , it changes all the time

2.Through APIs, we can request data from database;get the response from application

3.Example of well-known APIs:

    Google Maps API
    Facebook API
    Weather API
    Youtube API
    Twitter API
  
4.Elements of an API:

    Access
    Request:methods-->get;post;delet;patch;put
            parameters-->像是url的后缀一样，用于搜索的限制，帮助筛选出想要的东西
    Response
    
5.HTTP:

    5.1 In order to get data using API, weneed to make an API request call to retrieve data from any website or 
        its server
    5.2 like we do while visiting any website ,we enter an URL, URL is reference to a specitfic web page
    5.3 URL consists of 2 parts: 1)protocol identifier: http or https
                                 2)host name:amazon.in
    5.4 HTTP: hyper text transfer protocol
              1) protocol
                 1.1) protocols are nothing but a set of rules which defines how to format and transmit the 
                      request to server, and how to receive respinse from server.
                 1.2) A network protocol defines rules and conventions for communication between network devices
              2) the communication between client(web browser) and server on internet happens using HTTP
              3) Information is exchanged in the form of hypertext documents
              4) A web browser may be the client and an application on a computer that hosts a web site may ne 
                 the server
              5) Method of HTTP:
                 5.1) GET-->read(can not change)
                 5.2) POST-->write(can change)
    5.5 HTTP libraries:
        1) httplib
        2) urllib
        3) requests
    5.6 requests:
        1) it is an HTTP library written in Python
        2) it allows us to send HTTP requests using our python code and we can access the response also
    5.7 response object attribute
        1) status_code: 2xx -->200 request call is successful
                        4xx -->404 request call is failed
        2) encoding
        3) url
        4) text
        5) headers

### 2. JSON file format

    1.JSON - JavaScript Object Notation

    2.The most common formats for information retrieved, in APIs are JSON and XML

    3.It is a lightweight,text-based,human-readable file format, and can be edited using a text editor

    4.It has 2 parts: key / value

    5.JSON data types:
  
          Primitive data types: number/string/boolean/null
          Complex data types:Array/Object

### 3. JSON to python

    JSON--------------->Python
    object              dict
    number              int,float
    boolean             boolean
    string              str
    array               list/tuple
    null                None

```python
import json
```

#### 3.1 json object--->python dict

```python
json_data='{"student":"Mohit"}'
print(type(json_data))
python_data=json.loads(json_data)
print(python_data)
print(type(python_data))
print(python_data['student'])
```

```python
json_data='{"rollnumber":101}'
python_data=json.loads(json_data)
print(python_data)
print(python_data['rollnumber'])
```

```python
json_data='{"student":{"name":"mohit","rollnumber":101}}'
python_data=json.loads(json_data)
print(python_data)
student_details=python_data['student']
print(student_details)
name=student_details['name']
rollnumber=student_details['rollnumber']
print(name)
print(rollnumber)
```

#### 3.2 json array--> python list/tuple

```python
json_data='{"fruits":["apple","orange","banana"]}'
python_data=json.loads(json_data)
print(python_data)
print(type(python_data))
print(python_data['fruits'])
print(type(python_data['fruits']))

for i in python_data['fruits']:
    print(i)
```

### 6. POST request

```python
response=requests.post('http://httpbin.org/post')
print(response)
print(response.text)
```

```python
#requests.post('url',data)
#data refers to the data you want to update
response=requests.post('http://httpbin.org/post',data={'key1':"val1",
                                                       'key2':"val2"})
print(response)
print(response.text)
```

```python
#pass values -->method1
response=requests.post('http://httpbin.org/post',data={'key1':["val1","val3"],
                                                       'key2':"val2"})
print(response)
print(response.text)
```

```python
#pass values -->method2
response=requests.post('http://httpbin.org/post',data=[('key1',"val1"),('key1',"val3"),
                                                       ('key2',"val2")])
print(response)
print(response.text)
```

```python
response=requests.post('https://jsonplaceholder.typicode.com/posts')
print(response)
print(response.text)
```

```python
response=requests.post('https://jsonplaceholder.typicode.com/posts',data={'userId':2, id:200, "title":"test"})
print(response)
print(response.text)
```
